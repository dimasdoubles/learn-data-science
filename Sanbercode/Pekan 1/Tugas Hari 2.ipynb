<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?

Function adalah blok kode yang melakukan tugas tertentu dan menghasilkan sebuah return.
Method adalah function yang terintegrasi ke suatu object. Sedikit gambaran mengenai method, misal kita punya object ikan dan burung. Untuk bergerak, ikan punya function "berenang", burung punya function "terbang". Fungsi berenang hanya bisa digunakan untuk menggerakan ikan, dan fungsi terbang hanya bisa digunakan untuk menggerakan burung.

<h2>Soal 4: Menggunkan Method String</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    return sum(inp_list) / len(inp_list)
    
print(mean_list(obj_list))
```

Expected Output:

13.9

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(a_list, b_list):
    return a_list + b_list
    
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
