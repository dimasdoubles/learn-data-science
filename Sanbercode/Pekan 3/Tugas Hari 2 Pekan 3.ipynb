Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

# membuat figure dan axis
fig, ax = plt.subplots()

# mengubah axis label dan title
ax.set_xlabel("X-Axis Label")
ax.set_ylabel("Y-Axis Label")
ax.set_title("Grafik Cos Sin")


plt.plot(x, cos, color="r", label="Cosx")
plt.plot(x, cos3, color="g", linestyle="dashed", label="3 Cosx")
plt.plot(x, sin, color="b", linestyle="-.", label="Sin(x-90)")
plt.plot(x, sin2, color="y", marker="x", linestyle="none", label="2 Sinx")
plt.legend(loc="center")
plt.show()
```
