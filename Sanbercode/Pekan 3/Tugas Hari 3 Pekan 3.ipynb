```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, sharex = True, sharey=True)
plot_tan, = ax[0].plot(x, tan, color="b")
plot_sin, = ax[1].plot(x, sin, color="r")
plot_cos, = ax[2].plot(x, cos, color="g")

ax[0].set_title("Grafik tanx")
ax[1].set_title("Grafik sinx")
ax[2].set_title("Grafik cosx")
plt.xticks([-6, -4, -2, 0, 2, 4, 6])
fig.legend((plot_tan, plot_sin, plot_cos), ("tan X", "sin X", "cos X"), "upper center")

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(8, 7))
ax.plot(x, y)
ax.set_title("Grafik Cosx")

ins_kanan_atas = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins_kanan_atas.set_title("Grafik Cosx^4")
ins_kanan_atas.plot(x, y, color="b")
ins_kanan_atas.plot(x, y4, color="black")

ins_kanan_bawah = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins_kanan_bawah.set_title("Grafik Cosx^3")
ins_kanan_bawah.plot(x, y, color="b")
ins_kanan_bawah.plot(x, y3, color="g")

ins_kiri_atas = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins_kiri_atas.set_title("Grafik Cosx^5")
ins_kiri_atas.plot(x, y, color="b")
ins_kiri_atas.plot(x, y5, color="y")

ins_kiri_bawah = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins_kiri_bawah.set_title("Grafik Cosx^2")
ins_kiri_bawah.plot(x, y, color="b")
ins_kiri_bawah.plot(x, y2, color="r")

plt.show()
```
