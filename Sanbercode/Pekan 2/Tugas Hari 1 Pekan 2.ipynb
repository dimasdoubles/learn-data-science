<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 

list dan dictionary merupakan tipe data bawaan python, sedangkan Series merupakan Objek pandas. Series harus berisi data2 dengan tipe yang sama, list dan dictionary bisa berisi data dengan tipe yang berbeda.

Series adalah objek 1 dimensi, DataFrame adalah objek 2 dimensi.

```python
# code here
mangga = pd.Series(["Manalagi", "Gadung", "Arum manis", "Apel"])
print(mangga)
```
